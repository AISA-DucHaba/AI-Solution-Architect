# üêê Deploy using Gradio and Fastai

```python
import gradio as gr
import fastai
from fastai.vision.all import load_learner, PILImage
import json # Import the json library

# load learner
fname = './butterfly_learner_1762396100.pkl'
learn = load_learner(fname)

def classify_butterfly(image):
  pred, pred_idx, probs = learn.predict(PILImage(image))
  # Create a dictionary with the results
  results = {
      "prediction": str(pred),  # Convert prediction to string
      "prediction_index": int(pred_idx), # Convert prediction index to int
      "probabilities": probs.tolist() # Convert probabilities to a list
  }
  # Return the results as a JSON string
  return json.dumps(results, indent=2)

#
# Gradio
cnn_butterfly = gr.Interface(fn=classify_butterfly,
  inputs=gr.Image(type="pil"),
  outputs=gr.Json(label="Prediction (JSON)"), # Change output type to Json
  title="Butterfly CNN Tester",
  description="Upload a butterfly image to get the model‚Äôs prediction and class probabilities in JSON format.",
)
#
# lauch it
cnn_butterfly.launch()
```

```python
import pandas as pd

def get_random_sample(dataframe, num_samples=20):
  """
  Randomly selects a specified number of rows from a pandas DataFrame.

  Args:
    dataframe (pd.DataFrame): The input DataFrame.
    num_samples (int): The number of random samples to select. Defaults to 20.

  Returns:
    pd.DataFrame: A new DataFrame containing the randomly selected rows.
                  Returns an empty DataFrame if the input DataFrame has fewer
                  rows than num_samples.
  """
  if len(dataframe) < num_samples:
      print(f"Warning: The DataFrame has fewer than {num_samples} rows. Returning all rows.")
      return dataframe.copy() # Return a copy to avoid modifying the original DataFrame

  return dataframe.sample(n=num_samples, random_state=42) # Using a random_state for reproducibility

# Example usage (assuming df_training is your DataFrame)
random_sample_df = get_random_sample(df_training, 20)
random_sample_df.head()
```

```python
import os
import shutil
import pandas as pd

def export_sample_with_images(dataframe, output_dir="butterfly_test20", base_image_path="/content/butterfly_dataset"):
    """
    Exports a DataFrame and associated images to a new directory.

    Args:
        dataframe (pd.DataFrame): The input DataFrame with a 'filepaths' column
                                  containing paths relative to base_image_path.
        output_dir (str): The name of the directory to create for the exported data.
        base_image_path (str): The base directory where the images referenced
                               in the 'filepaths' column are located.
    """
    # 1. Create the new directory
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)
        print(f"Created directory: {output_dir}")
    else:
        print(f"Directory already exists: {output_dir}")

    # 2. Save the DataFrame
    csv_output_path = os.path.join(output_dir, "random_sample.csv")
    dataframe.to_csv(csv_output_path, index=False)
    print(f"DataFrame saved to: {csv_output_path}")

    # 3. Copy the images
    print(f"Copying images to {output_dir}...")
    for index, row in dataframe.iterrows():
        image_relative_path = row['filepaths']
        # Construct the full source path
        source_image_path = os.path.join(base_image_path, image_relative_path)

        # Construct the destination path, preserving the relative directory structure
        destination_image_path = os.path.join(output_dir, image_relative_path)

        # Create parent directories if they don't exist
        os.makedirs(os.path.dirname(destination_image_path), exist_ok=True)

        try:
            shutil.copy(source_image_path, destination_image_path)
            # print(f"Copied: {source_image_path} to {destination_image_path}")
        except FileNotFoundError:
            print(f"Warning: Image not found at {source_image_path}. Skipping.")
        except Exception as e:
            print(f"Error copying {source_image_path}: {e}")
    print("Image copying complete.")

# Example usage (assuming random_sample_df is your DataFrame)
export_sample_with_images(random_sample_df)
```

```python
import requests
import pandas as pd
import os

# Define the API endpoint
API_URL = "https://huggingface.co/duchaba/butterfly_predict"

# Define the function for the agentic testing
def agentic_test_butterfly_model(dataframe, base_image_path="/content/butterfly_dataset"):
    """
    Performs agentic testing on the butterfly classification model API.

    Args:
        dataframe (pd.DataFrame): DataFrame with 'filepaths' and 'labels' columns.
        base_image_path (str): Base path to the image files.

    Returns:
        list: A list of dictionaries containing the test results for each image.
    """
    test_results = []

    for index, row in dataframe.iterrows():
        # 1. Precede Input
        image_relative_path = row['filepaths']
        true_label = row['labels']
        full_image_path = os.path.join(base_image_path, image_relative_path)

        # Construct the URL to access the image directly if needed,
        # or prepare the image data for the API call.
        # For this API, it seems we need to send the image file itself.

        if not os.path.exists(full_image_path):
            print(f"Warning: Image not found at {full_image_path}. Skipping.")
            continue

        print(f"Testing image: {image_relative_path}")

        # 2. Decide Action & 3. Act
        try:
            with open(full_image_path, "rb") as f:
                image_data = f.read()

            response = requests.post(API_URL, data=image_data)
            response.raise_for_status() # Raise an exception for bad status codes
            api_output = response.json()

        except requests.exceptions.RequestException as e:
            print(f"Error during API call for {image_relative_path}: {e}")
            test_results.append({
                "image_path": image_relative_path,
                "true_label": true_label,
                "predicted_label": "API Error",
                "probabilities": None,
                "is_correct": False,
                "error": str(e)
            })
            continue
        except json.JSONDecodeError:
            print(f"Error decoding JSON response for {image_image_relative_path}")
            test_results.append({
                "image_path": image_relative_path,
                "true_label": true_label,
                "predicted_label": "JSON Decode Error",
                "probabilities": None,
                "is_correct": False,
                "error": "Invalid JSON response"
            })
            continue


        # 4. Reflect
        predicted_label = api_output.get("prediction")
        probabilities = api_output.get("probabilities")

        is_correct = (predicted_label == true_label)

        print(f" True Label: {true_label}, Predicted Label: {predicted_label}, Correct: {is_correct}")

        # 5. Output
        test_results.append({
            "image_path": image_relative_path,
            "true_label": true_label,
            "predicted_label": predicted_label,
            "probabilities": probabilities,
            "is_correct": is_correct
        })

    return test_results

# Example usage (assuming random_sample_df is available)
# if 'random_sample_df' in globals():
#     test_results = agentic_test_butterfly_model(random_sample_df)
# else:
#     print("random_sample_df not found. Please run the previous steps to create it.")

# You can then process and display the test_results
# For example, to calculate accuracy:
# correct_predictions = sum(result['is_correct'] for result in test_results)
# accuracy = (correct_predictions / len(test_results)) * 100 if test_results else 0
# print(f"\nOverall Accuracy: {accuracy:.2f}%")

# To view the detailed results:
# results_df = pd.DataFrame(test_results)
# display(results_df)
```
