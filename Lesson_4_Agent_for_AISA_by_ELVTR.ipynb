# Reading email (or gmail)

````python
# prompt: Write a python function with documentation for reading email.
#
import imaplib
import email
from email.header import decode_header

def fetch_latest_emails(email_address, app_password, mailbox="inbox", max_emails=5):
    """
    Connects to Gmail using IMAP and fetches the latest emails.

    Args:
        email_address (str): Your Gmail address (e.g., 'duc.haba@gmail.com').
        app_password (str): Gmail App Password generated via your Google Account.
        mailbox (str): Mailbox to select (default is 'inbox').
        max_emails (int): Maximum number of latest emails to retrieve.

    Returns:
        list of dict: Each dictionary contains 'subject', 'from', and 'body' of an email.
    """
    # Connect to the Gmail IMAP server
    imap = imaplib.IMAP4_SSL("imap.gmail.com")

    try:
        # Log in to the server
        imap.login(email_address, app_password)

        # Select mailbox
        imap.select(mailbox)

        # Search for all emails in the mailbox
        status, messages = imap.search(None, "ALL")
        email_ids = messages[0].split()

        # Get the latest emails
        fetched_emails = []
        for i in email_ids[-max_emails:]:
            res, msg = imap.fetch(i, "(RFC822)")
            for response in msg:
                if isinstance(response, tuple):
                    # Parse email
                    msg_obj = email.message_from_bytes(response[1])
                    subject, encoding = decode_header(msg_obj["Subject"])[0]
                    if isinstance(subject, bytes):
                        subject = subject.decode(encoding or "utf-8")

                    from_ = msg_obj.get("From")

                    # Extract the email body
                    if msg_obj.is_multipart():
                        for part in msg_obj.walk():
                            content_type = part.get_content_type()
                            content_disposition = str(part.get("Content-Disposition"))
                            if "attachment" not in content_disposition:
                                body = part.get_payload(decode=True)
                                if body:
                                    body = body.decode(errors="ignore")
                                    break
                    else:
                        body = msg_obj.get_payload(decode=True).decode(errors="ignore")

                    fetched_emails.append({
                        "subject": subject,
                        "from": from_,
                        "body": body.strip()
                    })

        return fetched_emails

    finally:
        imap.logout()
````

````python
# prompt: Rewrite the following "Original code" using Python Agent Langchain framework
#
# pip install langchain langchain-openai
from langchain_openai import ChatOpenAI
from langchain.prompts import PromptTemplate
from langchain.schema import StrOutputParser, RunnablePassthrough
from langchain.runnables import chain

# â”€â”€ 1.  Configure the LLM â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
llm = ChatOpenAI(
    model="gpt-4",
    temperature=0.0,
    api_key="YOUR-KEY-GOES-HERE",   # <â€” replace with real key
)

# â”€â”€ 2.  Build the prompt template â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
BASE_PROMPT = """
You are an email assistant.

A user says: "{query}"

If the input is toxic text (harassment, hateful, sexual, or violent),
reply with:
WARNING TOXIC: <reason> (<confidence>%)

Otherwise, reply with a helpful answer.
"""

prompt = PromptTemplate.from_template(BASE_PROMPT)

# â”€â”€ 3.  Assemble the LangChain 'graph' â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
@chain
def toxicity_router(reply: str) -> str:
    """Post-process the LLM reply and apply side-effects / routing."""
    if reply.startswith("WARNING TOXIC:"):
        print(f"{'-> WARNING TOXIC:':>30} Use FBI API to report this: {reply}")
    else:
        print(f"{'-> Response:':>30} {reply}")
    return reply  # Return in case caller wants the text programmatically.

pipeline = (
    {"query": RunnablePassthrough()}     # feed raw user text into the prompt
    | prompt                             # fill the template
    | llm                                # call GPT-4
    | StrOutputParser()                  # get plain string back
    | toxicity_router                    # custom routing logic
)

# â”€â”€ 4.  Example runs â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
if __name__ == "__main__":
    pipeline.invoke("What is the capital of France?")
    pipeline.invoke("I like to eat pizza while killing you.")
````

````python
# prompt: Rewrite the following "Original code" using Python Agent LLamaIndex framework
from llama_index.core.llms import OpenAI
from llama_index.agent.openai import OpenAIAgentWorker
from llama_index.core.agent import AgentRunner
from llama_index.core.prompts import PromptTemplate
import os

# â”€â”€ Step 1: Set up API key â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
os.environ["OPENAI_API_KEY"] = "YOUR-KEY-GOES-HERE"  # replace with your actual key

llm = OpenAI(model="gpt-4", temperature=0.0)

# â”€â”€ Step 2: Define the base prompt â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
prompt_template_str = """
You are an email assistant.

A user says: "{user_input}"

If the input is toxic text (harassment, hateful, sexual, or violent),
respond with:
WARNING TOXIC: <reason> (<confidence>%)

Otherwise, reply with a helpful answer.
"""

prompt_template = PromptTemplate(template=prompt_template_str)

# â”€â”€ Step 3: Create an AgentWorker with the prompt â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
worker = OpenAIAgentWorker(
    llm=llm,
    system_prompt=prompt_template
)

agent = AgentRunner(worker)

# â”€â”€ Step 4: Define agent loop â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
def agent_loop_llamaindex(user_input):
    response = agent.chat({"user_input": user_input})
    reply = str(response)

    print(f'\n{"Agent Reply: ":>30} {reply}')

    if "WARNING TOXIC:" in reply:
        print(f'{"-> WARNING TOXIC:":>30} Use FBI API to report this: {reply}')
    else:
        print(f'{"-> Response:":>30} {reply}')

# â”€â”€ Step 5: Run examples â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
if __name__ == "__main__":
    agent_loop_llamaindex("What is the capital of France?")
    agent_loop_llamaindex("I like to eat pizza while killing you.")
````

````python
# prompt: Rewrite the following "Original code" using Python Agent AutoGen (Microsoft) framework
from autogen import AssistantAgent, UserProxyAgent, config_list_from_json
import os

# â”€â”€ Step 1: Configuration â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
# Set your OpenAI API key
os.environ["OPENAI_API_KEY"] = "YOUR-KEY-GOES-HERE"

# Optional: Use config from JSON if you want more structured setup
# config_list = config_list_from_json("OAI_CONFIG_LIST.json")

# â”€â”€ Step 2: Define the assistant agent â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
email_assistant = AssistantAgent(
    name="EmailAssistant",
    llm_config={
        "config_list": [
            {
                "model": "gpt-4",
                "api_key": os.environ["OPENAI_API_KEY"],
            }
        ],
        "temperature": 0,
    },
    system_message=(
        "You are an email assistant.\n"
        "If the user input contains toxic text (harassment, hateful, sexual, or violent), "
        'respond with: WARNING TOXIC: <reason> (<confidence>%)\n'
        "Otherwise, give a helpful response."
    ),
)

# â”€â”€ Step 3: Define the user proxy logic â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
def agent_loop_autogen(user_input: str):
    user_proxy = UserProxyAgent(name="User", human_input_mode="NEVER")

    response = user_proxy.initiate_chat(
        recipient=email_assistant,
        message=user_input,
        summary_method="reflection_with_llm",
    )

    final_reply = str(response.chat_history[-1]["content"])
    print(f'\n{"Agent Reply: ":>30} {final_reply}')

    if "WARNING TOXIC:" in final_reply:
        print(f'{"-> WARNING TOXIC:":>30} Use FBI API to report this: {final_reply}')
    else:
        print(f'{"-> Response:":>30} {final_reply}')


# â”€â”€ Step 4: Example Runs â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
if __name__ == "__main__":
    agent_loop_autogen("What is the capital of France?")
    agent_loop_autogen("I like to eat pizza while killing you.")
````
